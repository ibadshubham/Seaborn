```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mean=[0,0]
cov=[[0.2,0],[0,3]]
```

```python
x_axis, y_axis=np.random.multivariate_normal(mean,cov,size=40).T
```

```python
x_axis
y_axis
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis,shade=True,color='green')
```

```python
sns.kdeplot(x_axis,y_axis)
```

```python
sns.kdeplot(x_axis,y_axis,shade=True,cmap='RdBu')
```

```python
sns.kdeplot(x_axis,y_axis,cmap='RdBu')
```

```python
sns.kdeplot(x_axis,y_axis,n_levels=30)
```

```python
sns.kdeplot(y_axis)
sns.kdeplot(y_axis,bw=0.3)
```

```python
sns.kdeplot(y_axis,vertical=True)
```
