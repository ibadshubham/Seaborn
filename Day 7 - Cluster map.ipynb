```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flights= sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights =flights.pivot('month','year','passengers')
```

```python
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False)
```

```python
sns.clustermap(flights,row_cluster=False)
```

```python
sns.clustermap(flights,cmap='coolwarm')
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=1)
```

```python
sns.clustermap(flights,linewidth=1.5)
```

```python
sns.clustermap(flights,linewidth=1.5,figsize=(9,9))
```

```python
sns.clustermap(flights,standard_scale=1)
```

```python
sns.clustermap(flights,standard_scale=0)
```
