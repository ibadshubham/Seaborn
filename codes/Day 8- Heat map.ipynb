```python
normal= np.random.rand(12,15)

sns.heatmap(normal)
```

```python
sns.heatmap(normal, vmin=0,vmax=2)
```

```python
flights= sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights =  flights.pivot('month','year','passengers')
```

```python
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True)
```

```python
sns.heatmap(flights,annot=True,fmt='d')
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=1.0)
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=1.0,vmin=150,vmax=500)
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=1.0,cmap='RdBu')
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=1.0,cmap='summer')
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=3.0,cmap='winter')
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=3.0,cmap='winter',center=flights.loc['June',1958])
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=3.0,cmap='winter',cbar=False)
```
