```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.violinplot(x='total_bill',data=tips)
```

```python
sns.violinplot(x='size',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,palette='husl')
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex',data=tips,palette='RdBu',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='RdBu', inner='quartile')
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='RdBu', inner='stick')
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips,palette='RdBu', inner='stick',split=True)
```
